```python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#set 1

#generacion de datos
cats = ['Norte', 'Sur', 'Este', 'Oeste']
conteos = np.array([120, 90, 150, 110])

#generacion de la pizarra
fig, ax = plt.subplots()
ax.bar(cats, conteos, label="regiones * cantidad")



#generacion de las barras
ax.grid(True, alpha=0.4)
ax.set_xlabel('Categorías')
ax.set_ylabel('Conteos')
ax.set_title('Grafico de Barras de Conteos')
ax.legend()
plt.show()
```

```python
#set 2
n = 120
x = np.linspace(0, 100, n)
y = 2.0 * x + np.random.normal(0, 25, n)  # pendiente ~2 + ruido

fig = plt.figure(figsize = (6, 4))
ax = plt.gca()

ax.scatter(x, y, alpha = 0.9)
ax.grid(True)
ax.set_title('Grafico de Dispersion')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()


```

```python
#set 3
muestra = np.random.normal(loc=50, scale=10, size=1000) + np.random.exponential(scale=2, size=1000)*0.5


fig = plt.figure(figsize = (6, 4))
ax = plt.gca()

ax.hist(muestra, bins = 20, alpha = 0.9)


media = np.mean(muestra)
mediana = np.median(muestra)
ax.axvline(media, linestyle = '--', linewidth = 1)
ax.axvline(mediana, linestyle = ':', linewidth = 1)


ax.set_title("Histograma distribucion de valores")
ax.set_xlabel("Valores")
ax.set_ylabel("Frecuencia")
ax.grid(True, axis='y', linestyle='--', alpha = 0.4)

plt.show()
```

```python
#set 4
t = np.arange(1, 25)
tendencia = 5 + 1.2 * t                       # componente de tendencia (creciente)
estacional = 4 * np.sin(2 * np.pi * t / 12)  # componente estacional anual (periodo=12)
ruido = np.random.normal(0, 2, size=t.shape) # componente aleatoria (ruido)
serie = tendencia + estacional + ruido

fig = plt.figure(figsize = (6, 4))
ax = fig.gca()

ax.plot(t, serie, marker = 'o')

ax.set_title('Lineas: serie temporal con tendencia y estacionalidad')
ax.set_xlabel('Periodo')
ax.set_ylabel('Valor')
ax.grid(True, linestyle='--', alpha=0.4)

plt.show()
```

```python
#set 6
categorias = ['A', 'B', 'C', 'D']
valores = np.array([35, 25, 22, 18])

fig, ax = plt.subplots()
ax.pie(valores, labels=categorias, autopct='%1.1f%%', startangle=90)

ax.set_title('Grafico de Pastel')
ax.axis('equal')
ax.legend()
plt.show()
```

```python
fs = 500
T  = 2.0
t = np.linspace(0, T, int(fs*T), endpoint=False)
f1, f2 = 50, 120
signal = 1.0*np.sin(2*np.pi*f1*t) + 0.6*np.sin(2*np.pi*f2*t) + 0.3*np.random.normal(0,1,len(t))

Y = np.fft.rfft(signal)
freqs = np.fft.rfftfreq(len(t), d=1/fs)
amplitud = np.abs(Y) / len(t)

fig = plt.figure(figsize = (6, 4))
ax = plt.gca()

ax.plot(freqs, amplitud, label="picos")
ax.set_xlim(0,200)
ax.set_xlabel("Frecuencia (Hz)")
ax.set_ylabel("Amplitud")
ax.set_title("Analisis espectral")
ax.grid(True, linestyle='--', alpha = 0.4)
ax.legend()
plt.show()
```

```python
n = 60
X = np.random.uniform(10, 100, size=n)
Y = 1.5 * X + np.random.normal(0, 15, size=n)
S = np.random.uniform(50, 1500, size=n)

fig = plt.figure(figsize = (6, 4))
ax = plt.gca()

#escater con tamaños
ax.scatter(X, Y , s=S, alpha = 0.9)


ax.set_title("Burbujas: tercera variable en tamaño")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.grid(True, linestyle='--', alpha = 0.4)

plt.show()
```
