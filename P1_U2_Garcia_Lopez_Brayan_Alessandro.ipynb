```python
media_edad = alumnos['edad'].mean()
alumnos["edad"].map(lambda edad2: edad2 - media_edad )
```

```python
media_edad = alumnos['edad'].mean()
alumnos["edad"].map(lambda edad2: edad2 + media_edad )
```
